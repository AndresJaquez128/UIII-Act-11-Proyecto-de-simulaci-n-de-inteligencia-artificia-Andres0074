##Genera un programa en Python que simule inteligencia artificial utilizando diccionarios, listas de diccionarios, estructuras condicionales `if`, `if else`, `elif`, ciclo `while True`, ciclo `for` y funciones.

##El programa debe estar relacionado con vendedores y utilizar un diccionario llamado `vendedores` con los siguientes atributos:

* Cartera de clientes
* Cuota mensual
* Producto estrella
* Zona de ventas
* Comisión acumulada
* Vehículo asignado

#Requisitos:

* El diccionario debe iniciar vacío.
* Debe crearse una lista de diccionarios.
* Debe existir un menú interactivo.
* Debe permitir agregar, listar, analizar y eliminar vendedores.
* El listado debe mostrarse en forma de tabla bien alineada.
* El programa debe solicitar entrada de datos (key y values).
* Utilizar inteligencia artificial simulada mediante recomendaciones usando condicionales.
* No utilizar programación orientada a objetos.
* No utilizar `try`.
* El ejemplo debe ser simple y atractivo para alumnos de preparatoria.
* Agregar comentarios para separar secciones del programa.


```python
vendedores = []

# =========================================================
# AGREGAR VENDEDOR
# =========================================================

def agregar_vendedor():

    nombre = input("Ingrese nombre del vendedor: ")
    cartera = int(input("Ingrese cartera de clientes: "))
    cuota = float(input("Ingrese cuota mensual: "))
    producto = input("Ingrese producto estrella: ")
    zona = input("Ingrese zona de ventas: ")
    comision = float(input("Ingrese comisión acumulada: "))
    vehiculo = input("Ingrese vehículo asignado: ")

    datos = {

        "Nombre": nombre,
        "Cartera": cartera,
        "Cuota": cuota,
        "Producto": producto,
        "Zona": zona,
        "Comision": comision,
        "Vehiculo": vehiculo

    }

    vendedores.append(datos)

    print("\nVendedor agregado correctamente.")

# =========================================================
# LISTAR VENDEDORES
# =========================================================

def listar_vendedores():

    if vendedores == []:

        print("\nLa lista está vacía.")

    else:

        print("\n==================== LISTA DE VENDEDORES ====================")

        print("---------------------------------------------------------------------------------------------------------------")

        print("{:<15} {:<12} {:<12} {:<18} {:<15} {:<15} {:<15}".format(

            "NOMBRE",
            "CLIENTES",
            "CUOTA",
            "PRODUCTO",
            "ZONA",
            "COMISION",
            "VEHICULO"

        ))

        print("---------------------------------------------------------------------------------------------------------------")

        for datos in vendedores:

            print("{:<15} {:<12} {:<12} {:<18} {:<15} {:<15} {:<15}".format(

                datos["Nombre"],
                datos["Cartera"],
                datos["Cuota"],
                datos["Producto"],
                datos["Zona"],
                datos["Comision"],
                datos["Vehiculo"]

            ))

        print("---------------------------------------------------------------------------------------------------------------")

# =========================================================
# ANALIZAR VENDEDOR
# =========================================================

def analizar_vendedor():

    if vendedores == []:

        print("\nNo hay vendedores registrados.")

    else:

        buscar = input("\nIngrese nombre del vendedor: ")

        encontrado = False

        for datos in vendedores:

            if datos["Nombre"] == buscar:

                encontrado = True

                print("\n================ ANÁLISIS IA =================")

                print("Nombre:", datos["Nombre"])
                print("Zona:", datos["Zona"])
                print("Producto estrella:", datos["Producto"])

                if datos["Comision"] >= 50000:

                    print("\nLa IA detectó vendedor experto.")
                    print("- Excelente nivel de ventas")
                    print("- Alto rendimiento mensual")
                    print("- Recomendado para supervisión")

                elif datos["Comision"] >= 20000:

                    print("\nLa IA detectó vendedor intermedio.")
                    print("- Buen desempeño")
                    print("- Puede mejorar su cartera")

                else:

                    print("\nLa IA detectó vendedor principiante.")
                    print("- Debe aumentar ventas")
                    print("- Mejorar atención a clientes")

                if datos["Cartera"] >= 100:

                    print("- Tiene una cartera amplia de clientes.")

                else:

                    print("- Necesita conseguir más clientes.")

        if encontrado == False:

            print("\nVendedor no encontrado.")

# =========================================================
# ELIMINAR VENDEDOR
# =========================================================

def eliminar_vendedor():

    if vendedores == []:

        print("\nLa lista está vacía.")

    else:

        eliminar = input("\nIngrese nombre del vendedor a eliminar: ")

        eliminado = False

        for datos in vendedores:

            if datos["Nombre"] == eliminar:

                vendedores.remove(datos)

                eliminado = True

                print("\nVendedor eliminado correctamente.")

        if eliminado == False:

            print("\nVendedor no encontrado.")

# =========================================================
# MENÚ PRINCIPAL
# =========================================================

while True:

    print("\n====================================================")
    print("        IA PARA ADMINISTRAR VENDEDORES")
    print("====================================================")

    print("1. Agregar vendedor")
    print("2. Listar vendedores")
    print("3. Analizar vendedor")
    print("4. Eliminar vendedor")
    print("5. Salir")

    opcion = input("\nSeleccione una opción: ")

    # =====================================================
    # OPCIONES
    # =====================================================

    if opcion == "1":

        agregar_vendedor()

    elif opcion == "2":

        listar_vendedores()

    elif opcion == "3":

        analizar_vendedor()

    elif opcion == "4":

        eliminar_vendedor()

    elif opcion == "5":

        print("\nPrograma finalizado.")
        break

    else:

        print("\nOpción incorrecta.")
print('PROGRAMA REALIZADO POR JAQUEZ ANDRES 0074 NL = 27')
```
